### distribution of errors noise scaling

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
df = pd.read_csv('scaling_plots_u_bar_138.109_I_max_1.419.csv')

# Calculate ratios
u_bar_ratio = df['u_bar_error'] / df['fitted_u_bar'] 
I_max_ratio = df['I_max_error'] / df['fitted_I_max'] 

# Filter out values larger than 100
u_bar_ratio_filtered = u_bar_ratio[u_bar_ratio <= 100]
I_max_ratio_filtered = I_max_ratio[I_max_ratio <= 100]

# Print some statistics
print(f"u_bar ratio - Median: {u_bar_ratio_filtered.median():.2f}")
print(f"I_max ratio - Median: {I_max_ratio_filtered.median():.2f}")

# Create histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Histogram of fitted_u_bar to u_bar_error ratio
ax1.hist(u_bar_ratio_filtered, bins=30, alpha=0.7, edgecolor='black', 
         label=f'Median: {u_bar_ratio.median():.2f}')
ax1.set_xlabel('fitted_u_bar / u_bar_error')
ax1.set_ylabel('Frequency')
ax1.set_title('Distribution of fitted_u_bar to u_bar_error ratio < 100')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Histogram of fitted_I_max to I_max_error ratio
ax2.hist(I_max_ratio_filtered, bins=30, alpha=0.7, edgecolor='black',
         label=f'Median: {I_max_ratio.median():.2f}')
ax2.set_xlabel('fitted_I_max / I_max_error')
ax2.set_ylabel('Frequency')
ax2.set_title('Distribution of fitted_I_max to I_max_error ratio < 100')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()

# Print some statistics
print(f"u_bar ratio - Median: {u_bar_ratio_filtered.median():.2f}")
print(f"I_max ratio - Median: {I_max_ratio_filtered.median():.2f}")
```

### distribution of errors cell scaling

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the cell scaling data
df_cell = pd.read_csv('cell_scaling_N0_2328676.073_s_1.100_I_inf_3.141.csv')

# Calculate ratios for cell scaling parameters
N0_ratio = df_cell['N0_error'] / df_cell['N0'] 
s_ratio = df_cell['s_error'] / df_cell['s']
I_inf_ratio = df_cell['I_inf_error'] / df_cell['I_inf']

# Filter out infinite and NaN values, and limit to ratios up to 2
N0_ratio_finite = N0_ratio[np.isfinite(N0_ratio) & (N0_ratio <= 10)]
s_ratio_finite = s_ratio[np.isfinite(s_ratio) & (s_ratio <= 10)]
I_inf_ratio_finite = I_inf_ratio[np.isfinite(I_inf_ratio) & (I_inf_ratio <= 10)]

# Calculate statistics for the whole data (finite values)
N0_ratio_all_finite = N0_ratio[np.isfinite(N0_ratio)]
s_ratio_all_finite = s_ratio[np.isfinite(s_ratio)]
I_inf_ratio_all_finite = I_inf_ratio[np.isfinite(I_inf_ratio)]

# Create histograms
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

# Histogram of N0 to N0_error ratio
ax1.hist(N0_ratio_finite, bins=30, alpha=0.7, edgecolor='black')
ax1.set_xlabel('N0_error / N0')
ax1.set_ylabel('Frequency')
ax1.set_title(f'Distribution of N0_error to N0 ratio (2)\nMedian: {N0_ratio_all_finite.median():.2f}')
ax1.grid(True, alpha=0.3)

# Histogram of s to s_error ratio
ax2.hist(s_ratio_finite, bins=30, alpha=0.7, edgecolor='black')
ax2.set_xlabel('s_error / s')
ax2.set_ylabel('Frequency')
ax2.set_title(f'Distribution of s_error to s ratio (2)\nMedian: {s_ratio_all_finite.median():.2f}')
ax2.grid(True, alpha=0.3)

# Histogram of I_inf to I_inf_error ratio
ax3.hist(I_inf_ratio_finite, bins=30, alpha=0.7, edgecolor='black')
ax3.set_xlabel('I_inf_error / I_inf')
ax3.set_ylabel('Frequency')
ax3.set_title(f'Distribution of I_inf_error to I_inf ratio (2)\nMedian: {I_inf_ratio_all_finite.median():.2f}')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print some statistics
print(f"N0 ratio - Mean: {N0_ratio_finite.mean():.2f}, Median: {N0_ratio_finite.median():.2f}, Std: {N0_ratio_finite.std():.2f}")
print(f"s ratio - Mean: {s_ratio_finite.mean():.2f}, Median: {s_ratio_finite.median():.2f}, Std: {s_ratio_finite.std():.2f}")
print(f"I_inf ratio - Mean: {I_inf_ratio_finite.mean():.2f}, Median: {I_inf_ratio_finite.median():.2f}, Std: {I_inf_ratio_finite.std():.2f}")
```

### visualize the results with tightests fits

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Load the cell scaling results
cell_scaling_df = pd.read_csv('intermediate_cell_scaling_fits/cell_scaling_N0_9992824.691_s_8.377_I_inf_9.705.csv')

# Print number of rows in the dataframe
print(f"Number of rows in the dataframe: {len(cell_scaling_df)}")

# Extract avg_uncertainty_error column
avg_uncertainty_error = cell_scaling_df['avg_uncertainty_error']

# Remove any non-finite values and values larger than 100
avg_uncertainty_error_filtered = avg_uncertainty_error[np.isfinite(avg_uncertainty_error) & (avg_uncertainty_error <= 500)]

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(avg_uncertainty_error_filtered, bins=30, alpha=0.7, edgecolor='black')
plt.xlabel('Average Uncertainty Error')
plt.ylabel('Frequency')
plt.title(f'Distribution of Average Uncertainty Error (10)\nMedian: {avg_uncertainty_error_filtered.median():.3f}\n{len(avg_uncertainty_error_filtered)} out of {len(cell_scaling_df)} values shown')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Print statistics
print(f"Average Uncertainty Error Statistics:")
print(f"Mean: {avg_uncertainty_error_filtered.mean():.3f}")
print(f"Median: {avg_uncertainty_error_filtered.median():.3f}")
print(f"Std: {avg_uncertainty_error_filtered.std():.3f}")
print(f"Min: {avg_uncertainty_error_filtered.min():.3f}")
print(f"Max: {avg_uncertainty_error_filtered.max():.3f}")
print(f"Number of filtered values: {len(avg_uncertainty_error_filtered)} out of {len(cell_scaling_df)} total")
```
